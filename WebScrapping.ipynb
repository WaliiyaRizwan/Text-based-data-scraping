```python
import re
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
 def open_url(url):
    # Initialize the web driver
    options = webdriver.ChromeOptions()
    options.add_argument("start-maximized")
    options.add_argument("--disable-extensions")
    options.add_argument("--headless")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--no-sandbox")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option("useAutomationExtension", False)
    driver = webdriver.Chrome(options=options)

    # Load the webpage
    driver.get(url)
    # Wait for the dynamic content to be generated
    time.sleep(10)
    # Enable JavaScript
    driver.execute_script("return navigator.webdriver")

    # Use BeautifulSoup to parse the HTML
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    # Close the web driver
    driver.quit()
    
    return soup
```

```python
first_url = "https://www.bedbathandbeyond.com/store/category/bedding/duvet-covers/15507?removeInStock=true%22"
base_url = "https://www.bedbathandbeyond.com/store/category/bedding/duvet-covers/15507/"
suffix = "?removeInStock=true"
page_url = []
page_url.append(first_url)
for i in range(2, 41):
    new_url = base_url + str(i) + "-48" + suffix
    page_url.append(new_url)
    
    
page_url
```

```python
#extracting product urls on each page from each page and storing them in a list
def get_url(soup):
    product_url_list = []
    product_url_list_src = []
    product_url_list_src = soup.find_all('a',{'class','PrimaryLink_7JKP inline-block'})
    for val in product_url_list_src:
        try:
            product_url_list.append(val.get('href'))
        except:
            pass
    return product_url_list
```

```python
# Create an empty list to store the extracted product URLs
products_url = []

i = 0
# Iterate over the URLs in the page_url list
for url in page_url:
    try:
        # Open the URL and parse its content using BeautifulSoup
        soup = open_url(url)
        # Extract the product URLs from the parsed content and append them to the products_url list
        products_url.append(get_url(soup))
        i = i+1
        print("url" + " "+ str(i))

    except Exception as e:
        print(f"Error opening or parsing {url}: {e}")
```

```python
def get_product(soup):
    Product = soup.find_all('h1',{'class': 'block parent txtLt txt150 skuRollup prodTitle'})
    if(len(Product)>1):
        Product = Product[0].text
    else:
        Product = " "
    return Product
```

```python
def get_category(soup):
    Category = soup.find_all('a',{'class':'black breadcrumb2'})
    if(len(Category)>1):
        Category = Category[-1].text
    else:
        Category = " "
    return Category
```

```python
def get_brand(soup):
    Brand = soup.find_all('a',{'class':'bold uppercase txtUnder pdpBrandLink txt065 display'})
    if(len(Brand)>1):
        Brand = Brand.text
    else:
        Brand = " "
    return Brand
```

```python
def get_label(soup):
    Label = soup.find_all("div", {"class":"white absolute sliderBadge"})
    if (len(Label) > 1):
        Label = Label[0].text
    else:
        Label = " "
    return Label
```

```python
def get_old_price(soup):
   #extract old price, new price and selected colour
    Old_Price = soup.find_all("div",{"class":"inlineBlock gr025 trackIsPrice txt112 display lh12 bold"})
    if(len(Old_Price)>1):
        Old_Price = Old_Price[0].text
    else:
        Old_Price = " "
    return Old_Price
```

```python
def get_current_price(soup):
    Current_Price = soup.find_all("span",{"class":"bBlue noWrap txt112 display lh12"})
    if(len(Current_Price)>1):
        Current_Price = Current_Price[0].text
    else:
        Current_Price = " "
    return Current_Price
```

```python
def get_selected_color(soup):
    Selected_Color = soup.find_all("span",{"class":"colorActive txtSz"})
    if(len(Selected_Color)>1):
        Selected_Color = Selected_Color[1].text
        
    else:
        Selected_Color = " "
    return Selected_Color
```

```python
def get_color_types(soup):
    colors = []
    color_types = soup.find_all('button',{"class":"parent midCtr swatchTap btnColor"})
    if(len(color_types)>1):     
        for color in color_types:
            string = color.get('on')
            # Using regular expression to find the color value
            match = re.search(r"color:\s*'(\w+)'", string)
            if match:
                colors.append(match.group(1))
            
            #extract number of colors
        number_of_colors = len(colors)
        colors = pd.Series([colors])
    else:
        number_of_colors = "0"
        colors = " "
    
    return colors, number_of_colors
    
    
```

```python
def get_size(soup):
    tt = p_soup.find_all('div',{"aria-label":"Product size swatches"})
    my_list = []
    if(len(tt)>1):  
        for t in tt:
            my_list.append(t.text)
    
        my_list = [i.strip() for i in my_list if i.strip()]
        my_list = pd.Series([my_list])
    else:
        my_list = " "
        
    return my_list
```

```python
def get_about_product(soup):
    About_Product = soup.find_all('div',{"class":"vt15 parent descLimit"})
    if(len(About_Product)>1):
        About_Product = About_Product[0].text
    else:
        About_Product = " "

    return About_Product
```

```python
# features code needs improvement - see the excel provided

def get_features_specs(soup):
    key = p_soup.find_all("dt",{"class":"g0 v0 s6 bold featureKey"})
    key_list = []
        
    for f in key:
        key_list.append(f.text)
    
    keys =key_list[1:]

    Specifications_keys = keys[1:5]

    features_keys = keys[5:-1]
    
    val_code = p_soup.find_all("dd",{"class":"s6 g0 v0 featureVal"})
    val_list = []
        
    for f in val_code:
        val_list.append(f.text)
    
    vals = val_list[1:]

    features_vals = vals[5:-1]
    
    
    
    #Specs = {Specifications_keys[i]: Specifications_vals[i] for i in range(len(Specifications_keys))}
    feat = {features_keys[i]: features_vals[i] for i in range(len(features_keys))}
    
    # Convert dictionary to dataframe
    #s_df = pd.DataFrame.from_dict(Specs, orient='index')
    #s_df = s_df.transpose()

    # Convert dictionary to dataframe
    f_df = pd.DataFrame.from_dict(feat, orient='index')
    f_df = f_df.transpose()
    
    
    return f_df
```

```python
def get_recommendations(soup):
    Other_Recommendations = soup.find_all("div",{"class":"lineClamp2 sliderProdTitle noTap"})
    Recommendations = []
    if(len(Other_Recommendations)>1):
        for r in Other_Recommendations:
            Recommendations.append(r.text)
        
        #remove empty strings
        Recommendations = [i.strip() for i in Recommendations if i.strip()]
        Recommendations = pd.Series([Recommendations])
    else:
        Recommendations = " "
    
    return Recommendations
```

```python
def get_ratings(soup):
    Rating_code = soup.find_all("div",{"class":"s6 flex col vb1 ratingsAverage"})
    ratings = []
    for rate in Rating_code:
        ratings.append(rate.text)
    
    ratings = [i.strip() for i in ratings if i.strip()]

    if(len(ratings)>1):
        
        pattern = r'\d+\.\d+'
        match_num = re.search(pattern, ratings[0])

        if match_num:
            rating = float(match_num.group())
        else:
            rating = " "
    else:
        rating = " "
        
    return rating
```

```python
def get_details(soup,url):
    
    #extract details using functions above
    Product = get_product(soup)
    Category = get_category(soup)
    Brand = get_brand(soup)
    Label = get_label(soup)
    Reviews = get_reviews(soup)
    Old_Price = get_old_price(soup)
    Current_Price = get_current_price(soup) 
    Selected_Color = get_selected_color(soup)
    colors, number_of_colors = get_color_types(soup)
    size_list = get_size(soup)
    About_Product = get_about_product(soup)
    f_df = get_features_specs(soup)
    Recommendations = get_recommendations(soup)
    rating = get_ratings(soup)
    


    # make a dataframe
    df = pd.DataFrame()
    
    df['link'] = pd.Series(url)
    df['Category'] = pd.Series(Category)
    df['Brand'] = pd.Series(Brand)
    df['Name'] = pd.Series(Product)
    df['Label'] = Label
    df['Reviews'] = Reviews
    df["Rating"] = rating
    df["No. of Colours"] = number_of_colors
    df["Color Selected"] = Selected_Color
    df["Colour Names"] = colors
    df["All Sizes"] = size_list
    df["Current Price"] = Current_Price
    df["Old Price"] = Old_Price
    df["About the Product"] = About_Product
    df["Collection"]=Recommendations

    df = df.join(f_df)
    
    return df
```
